```python
import pandas as pd
import glob
```

```python
all_csvs = glob.glob('Raw/*.csv')
merged_df = pd.concat((pd.read_csv(f) for f in all_csvs),ignore_index=True)
merged_df.head()
```

```python
len(merged_df)
```

```python
merged_df.info()
merged_df.isnull().sum()
```

```python
#Filter for Northern California, Oregon, Washington
df_pnw = merged_df[
    merged_df['place'].str.contains("Oregon|Washington|OR|WA", case=False, na=False) |
    ((merged_df['place'].str.contains("California|CA", case=False, na=False)) & (merged_df['latitude'] >= 39.5))
]
# Display shape and a preview of the filtered dataset
df_pnw.head()
```
